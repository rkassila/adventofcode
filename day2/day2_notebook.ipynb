```python
url = 'data.txt'

with open(url, 'r') as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines]
```

```python
lines
```

```python
len(lines)
```

```python
for i in range(len(lines)):
    lines[i] = lines[i].split(':', 1)[-1].strip()

lines
```

```python
data = []
for line in lines:
    color_quantities = [pair.strip() for pair in line.split(';')]
    color_sum = {}
    for pair in color_quantities:
        colors = [c.strip() for c in pair.split(',')]
        for color in colors:
            quantity, color_name = color.split()
            quantity = int(quantity)
            color_sum[color_name] = color_sum.get(color_name, 0) + quantity
    data.append(color_sum)

data
```
