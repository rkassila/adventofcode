```python
url = 'text.txt'

with open(url, 'r') as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines]
```

```python
lines
```

```python
chiffres = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']

n_list = []

for line in lines:
    new_line = []
    i = 0

    while i < len(line):
        if line[i].isdigit():
            new_line.append(line[i])
            i += 1
        else:
            for chiffre in chiffres:
                if chiffre in line[i:i+len(chiffre)]:
                    new_line.append(str(chiffres.index(chiffre)))
                    i += 1
                    break
            else:
                i += 1

    n_list.append(new_line)

n_list
```

```python
new_new_list = []
for valeur in n_list:
    if len(valeur) < 2:
        new_new_list.append(valeur+valeur)
    else:
        new_new_list.append(valeur)


new_new_list
```

```python
new = []

for element in new_new_list:
    new.append(int(element[0]+element[-1]))

new
```
