```python
def CounterNumbers(a):
    numbers_with_symbols = []
    temp_i = 0
    temp_j = 0

    for i, line in enumerate(a):
        for j, char in enumerate(line):
            if char.isdigit():
                length = 1
                temp_num = char

                if j > 0:
                    if a[i][j-1].isdigit():
                        length += 1
                        temp_num = a[i][j-1] + temp_num

                        if j > 1:
                            if a[i][j-2].isdigit():
                                length += 1
                                temp_num = a[i][j-2] + temp_num

                if j < len(line) - 1:
                    if a[i][j+1].isdigit():
                        length += 1
                        temp_num = temp_num + a[i][j+1]

                        if j < len(line) - 2:
                            if a[i][j+2].isdigit():
                                length += 1
                                temp_num = temp_num + a[i][j+2]

                number = int(temp_num)

                adjacent_symbols = any(a[x][y] not in ".0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz" for x in range(max(0, i-2), min(len(lines), i+2)) for y in range(max(0, j-2), min(len(line), j+2)))

                if adjacent_symbols:
                    if len(numbers_with_symbols) >=1:
                        if temp_i != i or temp_j < j-2:
                            numbers_with_symbols.append(int(number))
                            temp_i = i
                            temp_j = j
                    else:
                        numbers_with_symbols.append(int(number))
                       #else:
                    #numbers_with_symbols.append(int(number))

    return numbers_with_symbols

CounterNumbers(lines)
```
