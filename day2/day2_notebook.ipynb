```python
url = 'data.txt'

with open(url, 'r') as file:
    lines = file.readlines()
    lines = [line.strip() for line in lines]

lines
```

```python
lines_r = []
for i in range(len(lines)):
    lines_r.append(lines[i].split(':', 1)[-1].strip())

lines_r
```

```python
new_lines = []

for line in lines_r:
    a = line.split(";")

    elo = []
    for elemen in a:
        color_dict = {}
        elements = elemen.split(",")
        for element in elements:
            value, key = element.strip().split()
            color_dict[key] = int(value)
        elo.append(color_dict)

    new_lines.append(elo)

new_lines
```

```python
ref = {'red' : 12, 'green' : 13, 'blue' : 14}

filtered_list = [sublist for sublist in new_lines if all(all(d[key] <= ref[key] for key in d) for d in sublist)]

filtered_list
```

```python
indexs = []

for id, item in enumerate(new_lines):
    if item in filtered_list:
        indexs.append(id + 1)

indexs
```

```python
sum(indexs)
```

```python
max_values_per_line = []

for line in new_lines:
    max_values = {color: max(entry.get(color, 0) for entry in line) for color in set(color for entry in line for color in entry)}
    max_values_per_line.append(max_values)

max_values_per_line
```

```python
len(max_values_per_line)
```

```python
from functools import reduce
from operator import mul

sum_lines = []

for line in max_values_per_line:
    result = reduce(mul, line.values())
    sum_lines.append(result)

sum_lines
```

```python
sum(sum_lines)
```
